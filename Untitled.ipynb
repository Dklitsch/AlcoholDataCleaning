Here we start cleaning the data. 

Rename the 'name' column from the world dataset so we can merge it with drinks, and remove continents that don't really have countries in them.

```python
#Rename name column to match the drinks dataset.

world = world.rename(columns = {'name':'country'})
world = world.loc[~world.continent.str.contains('Antarctica')]
world = world.loc[~world.continent.str.contains('Seven seas')]
world.head(5)
```

Fix the labels for countries so they match between datasets. The choice of which dataset to modify is completely arbitrary. I went with names that had the lease abrieviation in case we wanted to use this for another purpose later.

Merge the two datasets so we can get the geographic information from world and the alcohol information from drinks. Be sure to use a left merge so countries without rows in drinks don't disappear.

```python
world = world.merge(drinks, on='country', how='left')
```
